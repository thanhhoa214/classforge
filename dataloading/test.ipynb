# Import Required Libraries
Import pandas and any other necessary libraries such as os to handle file paths.

```python
import pandas as pd  # For handling Excel and CSV files
import os  # For handling file paths and directories
```

# Load Excel File and List Sheets
Use pandas to read the Excel file and list all sheet names with pd.ExcelFile().

```python
# Specify the path to the Excel file
excel_file_path = "data/Student Survey - Jan.xlsx"

# Load the Excel file
excel_file = pd.ExcelFile(excel_file_path)

# List all sheet names

sheet_names = excel_file.sheet_names
l = [sheetname.lower() for sheetname in sheet_names]
print("Sheets in the Excel file:", l)
```

# Convert Each Sheet to CSV
Iterate over each sheet, read it into a DataFrame, and convert it directly to CSV format.

```python
# Iterate over each sheet, read it into a DataFrame, and convert it directly to CSV format
output_directory = "data/csv"  # Directory to save the CSV files

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Loop through each sheet and save it as a CSV file
for sheet_name in sheet_names:
    # Read the sheet into a DataFrame
    df = excel_file.parse(sheet_name)
    
    df.columns = [col.strip().lower().replace(" ", "_").replace("-", "_") for col in df.columns]
    # Define the output CSV file path
    csv_file_path = os.path.join(output_directory, f"{sheet_name}.csv")
    
    # Save the DataFrame to a CSV file
    df.to_csv(csv_file_path, index=False)
    print(f"Saved sheet '{sheet_name}' to '{csv_file_path}'")
```

# Save CSV Files Using Sheet Names
Save each DataFrame to a CSV file, naming each file according to its corresponding sheet name.

```python
# Save CSV Files Using Sheet Names
# The code for this section is already included in the provided code snippet. It saves each DataFrame to a CSV file, naming each file according to its corresponding sheet name. Below is the same code for this section:

for sheet_name in sheet_names:
    # Read the sheet into a DataFrame
    df = excel_file.parse(sheet_name)
    
    # Define the output CSV file path
    csv_file_path = os.path.join(output_directory, f"{sheet_name}.csv")
    
    # Save the DataFrame to a CSV file
    df.to_csv(csv_file_path, index=False)
    print(f"Saved sheet '{sheet_name}' to '{csv_file_path}'")
```
